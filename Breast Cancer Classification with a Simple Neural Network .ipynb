```python
# Importing Librairies
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the dataset
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
#print(breast_cancer_dataset)
```

```python
# Loading the data into a dataframe
df=pd.DataFrame(breast_cancer_dataset.data, 
                columns=breast_cancer_dataset.feature_names)
df.head()
```

```python
# Adding the 'target' column to the dataframe
df['label']=breast_cancer_dataset.target
df.tail()
```

```python
# Getting some information about the data
df.info()
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Statistical mesures about the data 
df.describe()
```

```python
# Checking the distribution of Target variable 
df['label'].value_counts()
```

```python
# Groub by label
df.groupby('label').mean()
```

###### Separating the features and target

```python
# Separating the features and target
X=df.drop(columns='label', axis=1)
y=df['label']
```

```python
X.head()
```

##### Splitting the data into training data & Testing data

```python
# Splitting the data into training data & testing data
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2,
                                                 random_state=23)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

##### Building the NN

```python
# import tensorflow and keras 
import tensorflow as tf
tf.random.set_seed(3)# Handle reprocidible 
from tensorflow import keras # wrappers keras 
```

```python
# Setting up the layers of NN 
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),  # input layer, Flatten is the convert the data into single dimensional array
    keras.layers.Dense(20, activation='relu'),  # hidden layer, 2O neurons, with activaiton fonction relu
    keras.layers.Dense(2, activation='sigmoid')  # Output layer, 2 neurons equal to number class that we have as target variable
])
```

```python
# compiling the Neural Network 
model.compile(optimizer='adam', 
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# Training the Neural Network 
history=model.fit(X_train, y_train, validation_split=0.1, epochs=10)
```

As we can see the accuracy not the art of state let standard the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.fit_transform(X_test)
```

```python
# Training the Neural Network with Standardized values
history2=model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
```

###### Visualizing accuracy and loss

```python
# Plotting without Standardization
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy without standardization')
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
# Plotting without Standardization
plt.plot(history2.history['accuracy'])
plt.plot(history2.history['val_accuracy'])
plt.title('model accuracy with standardization')
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
# Plotting Validation loss 
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
plt.title('model loss with standardization')
plt.ylabel('loss')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc='lower right')
```

###### Accuracy of the model on test data

```python
loss, accuracy=model.evaluate(X_test_std, y_test)
```
